Q.1. Create two int type variables, apply addition, subtraction, division and multiplications
and store the results in variables. Then print the data in the following format by calling the
variables:
First variable is __ & second variable is __.
Addition: __ + __ = __
Subtraction: __ - __ = __
Multiplication: __ * __ = __
Division: __ / __ = __


Answer - 

Here's an example of how you can create two int variables, perform arithmetic operations on them, store the results in variables, and then print the data in the requested format:

# Create two int variables
first_variable = 5
second_variable = 3

# Perform arithmetic operations and store the results in variables
addition_result = first_variable + second_variable
subtraction_result = first_variable - second_variable
multiplication_result = first_variable * second_variable
division_result = first_variable / second_variable

# Print the data in the requested format
print("First variable is", first_variable, "& second variable is", second_variable)

print("Multiplication:", first_variable, "*", second_variable, "=", multiplication_result)

output : 

First variable is 5 & second variable is 3
Addition: 5 + 3 = 8
Subtraction: 5 - 3 = 2
Multiplication: 5 * 3 = 15
Division: 5 / 3 = 1.6666666666666667

Q.2. What is the difference between the following operators:
(i) ‘/’ & ‘//’
(ii) ‘**’ & ‘^’

Answer - 

(i) The operators '/' and '//' are both used for division, but they have different behaviors.

'/' is the regular division operator in Python and performs floating-point division. It returns a floating-point result, including decimal places if necessary. For example, 5 / 2 would give the result 2.5.

'//' is the floor division operator in Python. It performs integer division and returns the largest integer less than or equal to the division result. It discards the decimal part of the division result. For example, 5 // 2 would give the result 2.

Here's an example to illustrate the difference:


regular_division = 5 / 2
floor_division = 5 // 2

print("Regular Division:", regular_division)
print("Floor Division:", floor_division)


output:


Regular Division: 2.5
Floor Division: 2

(ii) The operators '**' and '^' are both used for exponentiation, but they have different meanings.

'**' is the exponentiation operator in Python. It raises a number to the power of another number. For example, 2 ** 3 would give the result 8, as 2 raised to the power of 3 is 8.

'^' is not a built-in operator for exponentiation in Python. In some other programming languages, like C, '^' may be used as an XOR (exclusive OR) operator for bitwise operations, but it is not used for exponentiation in Python.

Here's an example using the exponentiation operator '**':



exponentiation = 2 ** 3

print("Exponentiation:", exponentiation)

output:

Exponentiation: 8


To clarify, in Python, if you want to perform exponentiation, you should use the '**' operator, as '^' does not have the same meaning


Q.3.List the logical operators.

Answer - 

The logical operators in Python are:

and: The logical AND operator returns True if both operands are True, otherwise, it returns False. It evaluates to False as soon as one of the operands is False.

or: The logical OR operator returns True if at least one of the operands is True, otherwise, it returns False. It evaluates to True as soon as one of the operands is True.

not: The logical NOT operator is a unary operator that reverses the logical value of its operand. It returns True if the operand is False, and False if the operand is True.

These operators are typically used to combine and manipulate boolean values (True or False) in logical expressions or conditions.

Q.4. Explain right shift operator and left shift operator with examples.

Answer - 

The right shift operator (>>) and left shift operator (<<) are bitwise operators in Python. They are used to shift the bits of a binary number to the right or left, respectively. These operators can be used on integers to perform bitwise operations at the binary level.

Right Shift Operator (>>):
The right shift operator shifts the binary representation of a number to the right by a specified number of positions. It effectively divides the number by 2 raised to the power of the specified shift amount, discarding the rightmost bits.
Example:

number = 16  # Binary: 10000
shift_amount = 2

result = number >> shift_amount  # Shift right by 2 positions

print(result)  # Output: 4 (Binary: 100)


In this example, the number 16 is represented as 10000 in binary. Shifting it to the right by 2 positions using the right shift operator >> gives us the result 4, which is represented as 100 in binary.

Left Shift Operator (<<):
The left shift operator shifts the binary representation of a number to the left by a specified number of positions. It effectively multiplies the number by 2 raised to the power of the specified shift amount, appending zeros on the right.
Example:

number = 3  # Binary: 11
shift_amount = 2

result = number << shift_amount  # Shift left by 2 positions

print(result)  # Output: 12 (Binary: 1100)


In this example, the number 3 is represented as 11 in binary. Shifting it to the left by 2 positions using the left shift operator << gives us the result 12, which is represented as 1100 in binary.

Note that shifting beyond the available bits or beyond the limits of the data type may result in unexpected behavior or wrapping around. Additionally, shifting negative numbers with the right shift operator may introduce sign extension, where the sign bit is preserved.
