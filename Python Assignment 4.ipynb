Q.10. What is difference between the list methods append() and insert()?


Answer- The append() and insert() methods are used to add elements to a list in Python, but they differ in how the elements are added:

append() method:
The append() method is used to add an element at the end of a list. It takes a single argument, which is the element to be appended.

Example:

my_list = [1, 2, 3]
my_list.append(4)
print(my_list)


Output:

[1, 2, 3, 4]


In this example, append(4) adds the element 4 to the end of the list, modifying the list in-place.

insert() method:
The insert() method is used to insert an element at a specific position within a list. It takes two arguments: the index at which the element should be inserted and the element itself.

Example:

my_list = [1, 2, 3]
my_list.insert(1, 4)
print(my_list)


Output:

[1, 4, 2, 3]

Slicing: Both lists and strings support slicing, which allows you to extract a portion of the sequence by specifying a range of indices. Slicing is done using square brackets ([]) and the start:stop:step notation.

Immutability vs. Mutability: Strings are immutable, which means that once a string is created, it cannot be modified. Any operation that seems to modify a string actually creates a new string. Lists, on the other hand, are mutable, so you can modify their elements, add or remove elements, or change their length.

Despite these similarities, there are also important differences between lists and strings. Lists can store elements of different types and are mutable, whereas strings store a sequence of characters and are immutable. The available methods and operations for lists and strings are also specific to each type.







Q.13.What&#39;s the difference between tuples and lists?


Answer-

Tuples and lists are both used to store collections of elements in Python, but they have several key differences:

Mutability: Tuples are immutable, meaning their elements cannot be modified once created. In contrast, lists are mutable, allowing you to modify, add, or remove elements.

Syntax: Tuples are defined using parentheses (), while lists are defined using square brackets [].

Example:

my_tuple = (1, 2, 3)
my_list = [1, 2, 3]


Element Types: Tuples can contain elements of different data types, and the elements can be heterogeneous. Lists can also store elements of different types, but they are commonly used to store homogeneous elements.

Use Cases: Tuples are typically used to represent collections of related values where the order of the elements is significant, such as coordinates, database records, or immutable configurations. Lists are commonly used when the order of the elements matters and when you need to modify or extend the collection.

Performance: Tuples are generally more memory-efficient and have a slightly faster performance compared to lists. This is because tuples are immutable, allowing Python to optimize memory usage and certain operations.

Available Methods: Lists have a broader range of built-in methods compared to tuples. Lists provide methods like append(), insert(), remove(), and sort(), which modify the list in-place. Tuples have fewer methods and most of them are for basic operations like indexing and counting.

Example:

my_list.append(4)      # Modifies the list in-place
my_tuple.count(2)     # Counts the occurrences of a value


Considering these differences, you would choose between tuples and lists based on whether you need an ordered collection that should not be modified (tuple), or a mutable collection that allows modifications (list).

contain?


Answer- 

Variables that "contain" list values in Python do not directly store the list itself. Instead, they store a reference or pointer to the list object in memory.

In Python, variables are essentially names or labels that refer to objects in memory. When you assign a list value to a variable, the variable holds a reference to the memory location where the list object is stored. This means that the variable does not contain the list directly, but rather holds a reference to it.

Here's an example to illustrate this concept:


my_list = [1, 2, 3]


In this example, the variable my_list does not contain the list [1, 2, 3] itself. Instead, it holds a reference to the memory location where the list object is stored. Multiple variables can refer to the same list object, allowing you to have multiple references to the same list.

This reference-based behavior in Python is important to understand, as it affects how variables and their values behave when you manipulate or assign them. For example, if you assign a list variable to another variable, both variables will refer to the same list object in memory, and changes made through one variable will be reflected in the other.

Q.17.How do you distinguish between copy.copy() and copy.deepcopy()?


Answer-

The copy.copy() and copy.deepcopy() functions are part of the copy module in Python, and they are used to create copies of objects. However, they differ in terms of the level of copying performed:


copy.copy() (Shallow Copy):
The copy.copy() function creates a shallow copy of an object. It creates a new object and copies the references of the original object's elements to the new object. If the elements themselves are mutable (like lists or other objects), the shallow copy only creates new references to the original elements, not new copies of the elements.

Example:


import copy

original_list = [1, [2, 3], 4]
shallow_copy = copy.copy(original_list)

shallow_copy[1].append(5)

print(original_list)    # [1, [2, 3, 5], 4]
print(shallow_copy)     # [1, [2, 3, 5], 4]


In this example, modifying the nested list in the shallow copy also affects the original list because they share the same reference to the nested list.

copy.deepcopy() (Deep Copy):
The copy.deepcopy() function creates a deep copy of an object. It creates a new object and recursively copies all the elements and nested elements of the original object. This means that even if the elements themselves are mutable, the deep copy creates new copies of the elements, ensuring that modifications in one do not affect the other.

Example:


import copy

original_list = [1, [2, 3], 4]
deep_copy = copy.deepcopy(original_list)

deep_copy[1].append(5)

print(original_list)    # [1, [2, 3], 4]
print(deep_copy)        # [1, [2, 3, 5], 4]


In this example, modifying the nested list in the deep copy does not affect the original list because the deep copy created a new copy of the nested list.

In summary, copy.copy() performs a shallow copy, which creates a new object but shares references to nested elements, while copy.deepcopy() performs a deep copy, creating completely independent copies of all elements and nested elements. The choice between copy.copy() and copy.deepcopy() depends on whether you want to create a shallow or deep copy of the object.
